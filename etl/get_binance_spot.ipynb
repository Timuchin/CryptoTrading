```python
import os
import sys
# making sure that custom functions can be imported from parent folder
sys.path.append(os.path.dirname(os.getcwd()))

import pandas as pd
from binance.client import Client

from utils import getdata
```

```python
client = Client()
```

```python
start_date = '2019-01-01'
end_date = pd.Timestamp.today().strftime('%Y-%m-%d')
```

```python
tickers = client.get_all_tickers()
```

```python
symbols = []
for ticker in tickers:
    if ticker.get('symbol').endswith('USDT'):
        symbols.append(ticker.get('symbol'))
```

```python
spot_data = pd.DataFrame()
for symbol in symbols:
    try:
        tmp = getdata(client=client, symbol=symbol, interval='1d', start_time=start_date, end_time=end_date)
        tmp['symbol'] = symbol
        spot_data = pd.concat([spot_data, tmp], axis = 0)
    except:
        continue
```

```python
spot_data = spot_data.reset_index()
```

```python
spot_data.to_csv('../data/spot_data.csv', index=False)
```
